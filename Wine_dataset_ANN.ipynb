```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd




dataset = pd.read_csv('wine.csv',sep =";")
x = dataset.iloc[:, 0:11].values
y = dataset.iloc[:, 11].values

from keras.utils import np_utils
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
lb_y = LabelEncoder()
y = lb_y.fit_transform(y)
Y = np_utils.to_categorical(y)



# encoder = LabelEncoder()
# encoder.fit(Y)
# encoded_Y = encoder.transform(Y)
# # convert integers to dummy variables (i.e. one hot encoded)
# dummy_y = np_utils.to_categorical(encoded_Y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,Y,test_size = 0.4,random_state=0)

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense

classifier = Sequential()

# classifier.add(Dense(output_dim = 6, init = 'uniform', activation = 'relu', input_dim = 11))
               
# classifier.add(Dense(output_dim = 6, init = 'uniform', activation = 'relu'))
               
# classifier.add(Dense(output_dim =1, init = 'uniform', activation = 'softmax'))

classifier.add(Dense(6, input_dim=11, activation='relu'))
classifier.add(Dense(7, activation='softmax'))
```

```python
# classifier.compile(optimizer = 'adam', loss= 'categorical_crossentropy', metrics = ['accuracy'])

classifier.compile(optimizer = 'adam', loss= 'categorical_crossentropy', metrics = ['accuracy'])


```

```python
classifier.fit(x_train,y_train,batch_size = 32, nb_epoch = 200)
```

```python
y_pred = classifier.predict(x_test)
y_pred = (y_pred>0.5)

# for i in range(len(y_pred)):
#     if y_pred[i]<4:
#         y_pred = 3
#     elif y_pred[i]<5:
#         y_pred = 4
#     elif y_pred[i]<6:
#         y_pred = 5
#     elif y_pred[i]<7:
#         y_pred = 6
#     elif y_pred[i]<8:
#         y_pred = 7
#     else:
#         y_pred = 8
    

# from sklearn.metrics import confusion_matrix
# cm = confusion_matrix(y_pred,y_test)
# cm
```
