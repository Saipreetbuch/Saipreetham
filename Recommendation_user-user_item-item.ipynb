```python
import numpy as np
import math
def cosine_similarity(v1,v2):
    "compute cosine similarity of v1 to v2: (v1 dot v2)/{||v1||*||v2||)"
    sumxx, sumxy, sumyy = 0, 0, 0
    for i in range(len(v1)):
        x = v1[i]; y = v2[i]
        sumxx += x*x
        sumyy += y*y
        sumxy += x*y
    return sumxy/math.sqrt(sumxx*sumyy)

v1,v2 = [1,5,4,4,2,3,2], [5,3,4,2,1,4,3]
v3,v4 = [2,4,1,4,2], [5,2,4,1,1]
v5,v6 = [3,2,1,2], [3,1,2,4]
v7,v8 = [5,3,2,4,2,4,3,2], [4,3,1,4,1,1,5,1]
v9,v10 = [5,3,2,1,4,3,2], [2,4,4,3,4,5,1]
v11,v12 = [2,4,2,4], [1,1,4,1]
v13,v14 = [5,4,2,1,2], [4,3,2,2,2]
v15,v16 = [5,3,2,4,2,1,4,3,2], [5,1,4,2,4,4,1,2,3]
v17,v18 = [5,3,2,4,1], [4,2,5,3,4]
a = cosine_similarity(v1,v2)
b = cosine_similarity(v3,v4)
c = cosine_similarity(v5,v6)
d = cosine_similarity(v7,v8)
e = cosine_similarity(v9,v10)
f = cosine_similarity(v11,v12)
g = cosine_similarity(v13,v14)
h = cosine_similarity(v15,v16)
i = cosine_similarity(v17,v18)

u = [5,3,2,4,2,1,4,3,2]
v1 = [3,1,2,3,4,4]
v2 = [2,4,4,4,3,1,4,4,5,1]
v7 = [3,1,1,4,5,2,1]
v8 = [4,3,2,2,3,2]
v10 = [4,2,5,3,4,3,4]
u_avg = np.average(u)
v1_avg = np.average(v1)
v2_avg = np.average(v2)
v7_avg = np.average(v7)
v8_avg = np.average(v8)
v10_avg = np.average(v10)


s = (u_avg)+((4-v1_avg)*c+(4-v2_avg)*e+(2-v7_avg)*f+(3-v8_avg)*g+(4-v10_avg)*i)/(np.sum(c+e+f+g+i)) 

s
```
