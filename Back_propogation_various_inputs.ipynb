```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])
x = np.dstack(x1)[0]
t = np.array([[0],
              [1],
              [1],
              [0]])
y = np.dstack(t)[0]

#Random declaring of weights
W1= np.random.random((4,2)) - 0.5
W2= np.random.random((1,4)) - 0.5
b1 = np.zeros(4).reshape(4,1)
b2 = np.zeros(1).reshape(1,1)

loss=[]
itera=250
for k in range(0,itera):
    
    a = np.dot(W1,x)
    z1= a+b1
    a1 = 1.0/(1.0 + np.exp(-z1))
    c = np.dot(W2,a1)
    z2 = c+b2
    a2 = 1.0/(1.0 + np.exp(-z2))
    y1 = a2
    y2 = np.log(y1)
    y3 = np.log(1-y1)
    
 #Loss function   
    L = -np.sum(np.multiply(y,y2) + np.multiply(1-y,y3))/4
    loss.append(L)
    
#backpropogation
    dz2 = (a2-y)
    dw2 = np.dot(dz2,a1.T)
    db2 = dz2
    da1 = np.dot(W2.T,dz2)
    k = np.multiply(a1,(1-a1))
    dz1 = np.multiply(da1,k)    
    dw1 = np.dot(dz1,x.T)    
    db1 = dz1
    
    learning_rate = 0.5
    W1 = W1 - learning_rate * dw1
#    print('W1: ',W1)
    b1 = b1 - learning_rate * db1
#    print(b1.shape)
#    print('b1: ',b1)
    W2 = W2 - learning_rate * dw2
#    print('W2: ',W2)
    b2 = b2 - learning_rate * db2
#    print('b2: ',b2)

# print(a2)
# print(loss)
plt.plot([i for i in range(itera)],loss)
# plt.xlim(0,200)
# plt.ylim(0,1)
plt.xlabel('No.of Iterations')
plt.ylabel('Error')
plt.show()
```
